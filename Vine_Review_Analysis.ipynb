```python
# Import dependencies
import pandas as pd

# Read vine_table.csv into dataframe
file_path = "Resources/vine_table.csv"
vine_df = pd.read_csv(file_path)

vine_df.head()
```

```python
# Filter dataframe where total_votes >= 20
vine_filtered_df = vine_df[vine_df['total_votes'] >= 20]
vine_filtered_df.head()
```

```python
# Filter dataframe where helpful_votes / total_votes >= 0.5
filter_1 = ((vine_df['helpful_votes'] / vine_df['total_votes']) >= 0.5)
vine_fifty_df = vine_df[filter_1]
vine_fifty_df.head()
```

```python
# Filter vine_fifty_df where verified purchase is Y
filter_y = vine_fifty_df['verified_purchase'] == 'Y'
vine_y_df = vine_fifty_df[filter_y]
vine_y_df.head()
```

```python
# Filter vine_fifty_df where verified purchase is N
filter_n = vine_fifty_df['verified_purchase'] == 'N'
vine_n_df = vine_fifty_df[filter_n]
vine_n_df.head()
```

```python
#Determine the total number of reviews, the number of 5-star reviews, 
#and the percentage of 5-star reviews for the two types of review (paid vs unpaid)

# Paid Reviews Analysis
total_reviews = vine_fifty_df['total_votes'].count()
total_paid_reviews = vine_y_df['total_votes'].count()
five_stars_paid = len(vine_y_df[vine_y_df['star_rating']==5.0])
percentage_fifty_paid = round(float(five_stars_paid / total_reviews) * 100, 2)
print(f'Paid Reviews Analysis \n============================\nTotal Number of Paid Reviews: {total_paid_reviews}')
print(f'Total number of 5-Star Reviews: {five_stars_paid}')
print(f'Percentage of 5-Star Reviews: {percentage_fifty_paid}%')
```

```python
# Unpaid Reviews Analysis
total_unpaid_reviews = vine_n_df['total_votes'].count()
five_stars_unpaid = len(vine_n_df[vine_n_df['star_rating']==5.0])
percentage_fifty_unpaid = round(float(five_stars_unpaid / total_reviews) * 100, 2)
print(f'Unpaid Reviews Analysis \n============================\nTotal Number of Unpaid Reviews: {total_unpaid_reviews}')
print(f'Total number of 5-Star Reviews: {five_stars_unpaid}')
print(f'Percentage of 5-Star Reviews: {percentage_fifty_unpaid}%')
```
